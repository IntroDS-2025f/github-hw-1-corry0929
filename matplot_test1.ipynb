````python
# ‰Ω†ÁöÑÁ®ãÂºè

import numpy as np
import matplotlib.pyplot as plt

# Âõ∫ÂÆö‰∫ÇÊï∏Á®ÆÂ≠ê
np.random.seed(768)
n = 200

# ÂèÉËÄÉÁ®ãÂºèÔºöË≥áÊñôÁîüÊàêËàáËôïÁêÜ
x = np.random.normal(loc=2.0, scale=1.0, size=n)
y = x * 0.5 + np.random.normal(scale=0.8, size=n)
size_var = np.abs(np.random.normal(loc=20, scale=10, size=n))  # Âü∫Á§éÈªûÈù¢Á©ç(Ê≠£ÂÄº)

# Ë¶ñË¶∫ÂåñÁ∏ÆÊîæÔºàmatplotlib ÁöÑ s ‰ª£Ë°®„ÄåÈù¢Á©ç„ÄçÔºâ
size_scale = 10        # ÂèØÂæÆË™øÔºöËÆäÂ§ß‚ÜíÈªûÊõ¥Â§ß„ÄÅËÆäÂ∞è‚ÜíÈªûÊõ¥Â∞è
sizes = size_var * size_scale

# Áπ™Âúñ
fig, ax = plt.subplots()
sc = ax.scatter(x, y, c=y, s=sizes, alpha=0.7, cmap='viridis', edgecolors='none')

ax.set_title('Scatter with Colormap')  # title
ax.set_xlabel('x-values')
ax.set_ylabel('y-values')

cbar = plt.colorbar(sc, ax=ax)
cbar.set_label('y-values')             # colorbar label

ax.grid(True, alpha=0.3)

plt.show()

````
